```python
a = int(input("Enter a Number: "))
b = int(input("Enter another Number: "))
def add(a,b): return a+b
def sub(a,b): return a-b
def mult(a,b): return a*b
def div(a,b): return a/b if b != 0 else "Dont Enter 0 as denominator" #Ternary Operator
print("1. add \n 2. sub \n 3. multiply \n 4. Division")
choice = int(input("Enter your Choice: "))
switch = {
    1: add,
    2: sub,
    3: mult,
    4: div,
}
if choice in switch:
    # print("Result:", switch[choice](a,b))
    print(f"Result: {switch[choice](a,b)} ")
else:
    print("Invalid Choice")
```

```python
"""
When all the length of the sides of the triangle is known - a,b,c
Semi perimeter (s) - (a+b+c)/2
Area = square root of (s * (s-a) * (s-b) * (s-c))
"""

import math
a = float(input("Enter First Side of Triangle: "))
b = float(input("Enter a Second Side of Triangle: "))
c = float(input("Enter a Third Side of Triangle: "))

s = (a+b+c)/2
print(f"Perimeter of Triangle with sides {a}, {b}, {c} is {s} units.")

ar = s * (s - a) * (s - b) * (s - c)
area = math.sqrt(ar)
print(f"Area of the Triangle with Sides {a}, {b} and {c} is {area} units.")
print(f"Round of Area of the Triangle with Sides {a}, {b} and {c} is {round(area,2)} units.")

areaa = (s * (s - a) * (s - b) * (s - c)) ** 0.5
print("Area with different syntax", areaa)

areas = math.sqrt(s * (s - a) * (s - b) * (s - c))
print("Area with different syntax", areas)
```

```python
"""
Simple Interest = (P * R * T) / 100
P = Principal Amount
R = Rate of Interest
T = Time Duration
"""

principle_amount = float(input("Enter Principal Amount: "))
rate_of_interest = float(input("Enter Rate of Interest: "))
time_duration = float(input("Enter Time Duration: "))

simple_interest = (principle_amount * rate_of_interest * time_duration)/ 100
print(f"Simple Interest with Principal Amount: {principle_amount}, Rate of Interest: {rate_of_interest} and Time Duration: {time_duration} is {simple_interest}.")
```

```python
"""
Compund Interst
Amount = P(1+R/100) ** T
ci = Amount - P
P = Principal Amount
R = Rate of Interest
T = Time Duration
"""
principle_amount = float(input("Enter Principal Amount: "))
rate_of_interest = float(input("Enter Rate of Interest: "))
time_duration = float(input("Enter Time Duration: "))

Amount = principle_amount * ((1+(rate_of_interest/100)) ** time_duration)
Compound_Interest = Amount - principle_amount
print(f"Amount: {Amount}")
print(f"Compount Interest is: {Compound_Interest}")
```
